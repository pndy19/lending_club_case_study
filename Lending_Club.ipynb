#### Removing __Outliers__

```python
annual_inc_outlier_percentile = df_loan["annual_inc"].quantile(0.99)
df_loan = df_loan[df_loan["annual_inc"] < annual_inc_outlier_percentile]
df_loan["annual_inc"].describe()
```

```python
plt.figure(figsize=(15, 15))
plt.title('Annual income distribution')
df_loan['annual_inc'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

### Here, we can see that Annual Income has lot of values so we can't see it's distribution so we need to create ranges for it. Same goes for Interesr rate and Loan Amount. I am using "cut" function to create different bins.

```python
# Annual Income Range

bins = [0, 50000, 100000, 200000, 400000, 6000000]
slot = ['< 50k', '50k - 1L', '1L - 2L', '2L - 4L', '4L +']
df_loan['annual_inc_range'] = pd.cut(df_loan['annual_inc'], bins, labels=slot)
```

```python
df_loan['annual_inc_range'].unique()
```

```python
# Interest Rate Range

bins = [0, 5, 10, 15, 20, 25]
slot = ['0-5', '6-10', '11-15', '16-20', '21-25']
df_loan['int_rate_range'] = pd.cut(df_loan['int_rate'], bins, labels=slot)
```

```python
df_loan['int_rate_range'].unique()
```

```python
# Loan Amount Range

bins = [0, 5000, 10000, 15000, 20000, 25000,40000]
slot = ['0-5000', '5000-10000', '10000-15000', '15000-20000', '20000-25000','25000 and above']
df_loan['loan_amnt_range'] = pd.cut(df_loan['loan_amnt'], bins, labels=slot)
```

```python
df_loan['loan_amnt_range'].unique()
```
