

__Here, it has been observed that 'int_rate' and 'revol_util' have percentage value but coming in Categorical fields whereas
they should be Numerical. Also, 'term' having 'Months' in it which can be removed. Thus, doing required operation.__

```python
df_loan["term"] = df_loan.term.apply(lambda x: x.replace(' months',''))
```

```python
df_loan["int_rate"] = df_loan.int_rate.apply(lambda x: x.replace('%','')).astype(float)
```

```python
df_loan["revol_util"] = pd.Series(df_loan["revol_util"]).str.replace("%", "").astype(float)
```
