### Here, we can see that Annual Income has lot of values so we can't see it's distribution so we need to create ranges for it. Same goes for Interesr rate and Loan Amount. I am using "cut" function to create different bins.

```python
# Annual Income Range

bins = [0, 50000, 100000, 200000, 400000, 6000000]
slot = ['< 50k', '50k - 1L', '1L - 2L', '2L - 4L', '4L +']
df_loan['annual_inc_range'] = pd.cut(df_loan['annual_inc'], bins, labels=slot)
```

```python
def get_percentage_of_charged_off_by_column_cat(col):
    df2 = pd.DataFrame()
    for r in df_loan[col].unique():
        
        df2 = df2.append({col :r, 'percentage of Charged off' :round( len(df_loan[(df_loan['loan_status']=='Charged Off') & (df_loan[col]==r)])*100/len(df_loan[(df_loan['loan_status']!='Current') & (df_loan[col]==r)] ) ,2)
             ,},  ignore_index = True)
    return df2        
```
